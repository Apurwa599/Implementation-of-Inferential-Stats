```python
from statsmodels.stats.weightstats import ztest as ztest

# Enter IQ levels for 20 patterns
data = [88, 92, 34, 67, 98, 97, 97, 99, 99, 105, 109, 109, 109, 112, 112, 114, 115, 116, 117, 118]

# Perform the z-test
z_stat, p_value = ztest(data,value=100)
```

```python
print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import math
```

```python
np.random.seed(6)
```

```python
school_ages = stats.poisson.rvs(loc=18,mu=35,size=1500)
class_ages= stats.poisson.rvs(loc=18,mu=30,size=60)
```

```python
school_ages
```

```python
class_ages
```

```python
class_ages.mean()
```

```python
_,p_value=ttest_1samp(class_ages,popmean=school_ages.mean())
```

```python
school_ages.mean()
```

```python
if p_value<=0.05:
    print("rejejct H0")
else:
    print("")
```
